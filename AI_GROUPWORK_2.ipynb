```python
#From the South African movies, list the ones released in 2021
from datetime import datetime
import pandas as pd

data = {
    "Title": ["Slay", "Trippin with the Kandasamys", "I Am All Girls", "Blood & Water"],
    "Release_Year": [2021, 2021, 2021, 2020],
    "Date_Added": [
        datetime(2021, 4, 26),
        datetime(2021, 6, 4),
        datetime(2021, 5, 14),
        datetime(2020, 5, 20)
    ],
    "Rating": ["PG-13", "TV-14", "R", "TV-MA"],
    "Is_Original": [True, True, False, True]
}
df = pd.DataFrame(data)
movies_2021 = df[df["Release_Year"] == 2021]
print(movies_2021)
```

```python
#List the cast of movies released in 2021
movies_2021 = df[(df['type'] == 'Movie') & (df['release_year'] == 2021)]

cast_list_2021 = movies_2021['cast'].dropna().tolist()

all_actors = set()
for cast_string in cast_list_2021:
    actors = [actor.strip() for actor in cast_string.split(',')]
    all_actors.update(actors)

sorted_actors = sorted(list(all_actors))

# Display the full list
print(f"There are {len(sorted_actors)} unique actors in movies released in 2021. The full list is:")
print(sorted_actors)
```

```python
#QN1: Write a program that asks the user for a number of days then prints out the number of seconds.
days = int(input("Enter number of days: ")) #asks user for number of days
seconds = days * 24 * 60 * 60 #calculates number of seconds in the given days
print("Number of seconds in", days, "days is:", seconds)
```

```python
#QN2: Write a program that asks a user to input the radius then calculates the volume of a sphere.
# Use the exponential operator in python to compute (r^3).
import math
radius = float(input("Enter the radius of the sphere(cm): ")) #asks user for radius
volume = (4/3) * math.pi * radius ** 3
print("The volume of the sphere is:", volume)
```

```python
#QN3: Using functions, write a program should ask the user to enter a number corresponding to the side length of the square and display the area and perimeter of the square.
#Create a function that calculates the perimeter and area of a squarewhen given the side as input
def perimeter_area(side):
    area = side * side
    perimeter = 4 * side
    return area, perimeter

length = float(input("Enter the side length of the square: ")) #asks user for side length
area, perimeter = perimeter_area(length) #calls the function to act on the inputted value
print("Area of the square:", area) #Outputs the calculated area
print("Perimeter of the square:", perimeter) #Outputs the calculated perimeter
```

```python
#QN4: Write a program using functions that determines whether a character input by a user is uppercase or lower case.
#Function that checks the case of an input character
def case(char):
    if char.isupper():
        return "The character is uppercase"
    else:
        return "The character is lowercase"

character = input("Enter a character: ") #asks user for a character
result = case(character) #calls the function and stores the result
print(result) #Outputs the result
```

```python
#QN5: Write the program equivalent to the pseudocode
x = 0
y = 20

while y >= 6:
    y = y - 4
    x = x + 2/y
print("Final value of x is:", x)
```

```python
#QN6: Write a program that:
#Uses a loop for a user to continually input 5 values to populate an array
#Calculates and displays average of values input into the array
values = []
for x in range(5):
    value = int(input("Enter a value:"))
    values.append(value)

    avg = sum(values)/ len(values)

print("The average of all values input into the array is: ", avg)
```
